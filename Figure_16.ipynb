```python
#Importing required library
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.array([0, 1.6, 6.27, 15.8])
y1 = np.array([0, 0.6, 0.99, 1])

# Creating x axis with range and y axis with Sine
# Function for Plotting Sine Graph
x = np.arange(0, 5*np.pi, 0.1)
y = - 1/np.exp(0.4*x) + 2

# Plotting Sine Graph
plt.plot(x*5, y*100, color='royalblue', label = 'High resolution counter')
plt.plot(x1*5, 100*(y1+1), color = 'royalblue', linestyle = 'dashed', label = 'Low resolution counter')
plt.legend(loc = 'best')

plt.grid()
plt.xlabel('Time [s]', fontsize = 12)
plt.ylabel('Synaptic weight [%]', fontsize = 12)
plt.title('Potential overestimation of late-phase weight \n due to reduced counter resolution')
plt.show()
```
